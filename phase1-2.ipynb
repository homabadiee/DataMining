### Consistency

Rating and Rating Count are compared together; when Rating value is null, Rating Count must be null too and
when Rating Count is zero, Rating must also be zero. 

```python
# Rating & Rating Count
null_ratings_idx = df_ps.loc[pd.isna(df_ps['Rating']), :].index
null_ratings_cnt_idx = df_ps.loc[pd.isna(df_ps['Rating Count']), :].index
consistent_null = 0
for i in null_ratings_idx:
    for j in null_ratings_cnt_idx:
        if i == j:
            consistent_null += 1

inconsistent_nulls = len(null_ratings_idx) + len(null_ratings_cnt_idx) - 2 * (consistent_null)

zero_ratings_cnt_idx = df_ps.loc[(df_ps['Rating Count'] == 0), :].index
inconsistent_zero = 0
for i in zero_ratings_cnt_idx:
    if df_ps['Rating'][i] != 0:
        inconsistent_zero += 1


consistent1 = (record_cnt - inconsistent_nulls - inconsistent_zero) / record_cnt
```

Installs and Minimum Installs are compared together. Installs must be greater or equal than Minimum Installs.
